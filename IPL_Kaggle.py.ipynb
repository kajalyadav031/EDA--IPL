```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
ipl=pd.read_csv("matches.csv")
ipl.head()
```

```python
#Getting frequency of most man of the match award 
ipl['player_of_match'].value_counts()
```

```python
ipl['winner'].value_counts()
```

```python
#frequency of result
ipl['result'].value_counts()
```

```python
#No of toss win wrt to each team
ipl['toss_winner'].value_counts()
```

```python
#extracting records where team won batting first

batting_first=ipl[ipl['win_by_runs']!=0]
```

```python
batting_first.head()
```

```python
plt.figure(figsize=(5,5))
plt.hist(batting_first['win_by_runs'])
plt.show()
```

```python
#finding out the number of wins wrt to each win by after battinf first
batting_first['winner'].value_counts()
```

```python
#making a bar plot for top 3 teams with most win after batting first
plt.figure(figsize=(7,3))
plt.bar( list(batting_first['winner'].value_counts()[0:3].keys()),batting_first['winner'].value_counts()[0:3],color=['yellow','pink','purple'])
plt.show()
```

```python
plt.pie(batting_first['winner'].value_counts(),labels=list(batting_first['winner'].value_counts().keys()),autopct="%0.1f%%")
plt.show()
```

```python
#extracting those records where a team has won after batting second
batting_second=ipl[ipl['win_by_wickets']!=0]
```

```python
ipl.head()
```

```python
#looking at the head
batting_second.head()
```

```python
#Making a histogram for frequency of win w.r.t to number of wickets
plt.figure(figsize=(5,5))
plt.hist(batting_second['win_by_wickets'],bins=30)
plt.show()
```

```python
#Finding out the frequency of number of win w.r.t to each team after batting second
batting_second["winner"].value_counts()
```

```python
#making a bar plot for top 3 teams with most wins after batting second
plt.bar(list(batting_second['winner'].value_counts()[0:3].keys()),batting_second['winner'].value_counts()[0:3],color=['red','violet','green'])
plt.show()
```

```python
#making a pie chart for distribution of most win after batting second

plt.pie(batting_second['winner'].value_counts(),labels=list(batting_second['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
